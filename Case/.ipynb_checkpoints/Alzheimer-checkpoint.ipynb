#### Vimos que não há dados nulos em nosso dataframe

#### Então, a ideia agora é identificar as variáveis numéricas que podem ser transformadas em categóricas

```python
df.columns 
```

#### Seguindo a documentação, podemos dizer que as seguintes variáveis podem ser transformadas em categóricas:

#### Gender Ethnicity EducationLevel Smoking  FamilyHistoryAlzheimers CardiovascularDisease Diabetes Depression HeadInjury Hypertension MemoryComplaints BehavioralProblems Confusion Disorientation PersonalityChanges DifficultyCompletingTasks Forgetfulness Diagnosis

```python
#transformando em categoricas
colunas_categoricas = ['Gender', 'Ethnicity', 'EducationLevel', 'Smoking', 'FamilyHistoryAlzheimers', 'CardiovascularDisease', 'Diabetes', 'Depression', 'HeadInjury', 'Hypertension', 'MemoryComplaints', 'BehavioralProblems', 'Confusion', 'Disorientation' ,'PersonalityChanges', 'DifficultyCompletingTasks', 'Forgetfulness' ,'Diagnosis']
df[colunas_categoricas] = df[colunas_categoricas].astype('category')
df_categoricas = df[colunas_categoricas].astype('category')
```

```python
df_categoricas
```

```python
#Separando as variáveis numéricas 

colunas_numericas = ['Age', 'BMI', 'AlcoholConsumption',
'PhysicalActivity', 'DietQuality', 'SleepQuality', 'SystolicBP', 
'DiastolicBP', 'CholesterolTotal', 'CholesterolLDL','CholesterolHDL', 'CholesterolTriglycerides', 'MMSE', 'FunctionalAssessment', 
'ADL']
df_numericas = df[colunas_numericas].copy()
```

```python
df_numericas.describe()
```

```python
#Analisando as variáveis numéricas

fig, axes = pltt.subplots(5, 3, figsize=(16, 20)) 

sns.histplot(data=df, x='Age', bins=35, kde=True, ax=axes[0, 0])
sns.histplot(data=df, x='BMI', bins=35, kde=True, ax=axes[0, 1])
sns.histplot(data=df, x='AlcoholConsumption', bins=35, kde=True, ax=axes[0, 2])
sns.histplot(data=df, x='PhysicalActivity', bins=35, kde=True, ax=axes[1, 0])
sns.histplot(data=df, x='DietQuality', bins=35, kde=True, ax=axes[1, 1])
sns.histplot(data=df, x='SleepQuality', bins=35, kde=True, ax=axes[1, 2])
sns.histplot(data=df, x='SystolicBP', bins=35, kde=True, ax=axes[2, 0])
sns.histplot(data=df, x='DiastolicBP', bins=35, kde=True, ax=axes[2, 1])
sns.histplot(data=df, x='CholesterolTotal', bins=35, kde=True, ax=axes[2, 2])
sns.histplot(data=df, x='CholesterolLDL', bins=35, kde=True, ax=axes[3, 0])
sns.histplot(data=df, x='CholesterolTriglycerides', bins=35, kde=True, ax=axes[3, 1])
sns.histplot(data=df, x='MMSE', bins=35, kde=True, ax=axes[3, 2])
sns.histplot(data=df, x='FunctionalAssessment', bins=35, kde=True, ax=axes[4, 0])
sns.histplot(data=df, x='ADL', bins=35, kde=True, ax=axes[4, 1])
sns.histplot(data=df, x='ADL', bins=35, kde=True, ax=axes[4, 2])


pltt.tight_layout()
pltt.show()
```

##### As variáveis numéricas estão bem distribuidas/balanceadas entre si, o que nos indica que, provavelmente, não há outliers 

```python
df_corr = df.drop(columns=['DoctorInCharge',])
corr = df_corr.corr()

plt.figure(figsize=(20,20))
sns.heatmap(corr, annot=True, fmt='.2f', square= True,
            linewidth=.5, cmap='coolwarm',vmin=-1, vmax=1)
plt.title('Correlation matrix', fontweight='bold')
plt.show()
```

#### Algumas observações:
##### A maioria das variáves tem pouca correlação com o diagnóstico. No entanto, as variáveis MMSE, FunctionalAssessment, MemoryComplains,
##### BehaivioralProblems e ADL possuem uma correlação maior com a variavel Diagnosis, quando comparada as outras variáveis
##### Logo, essas variáveis serão importantes na criação do nosso modelo
